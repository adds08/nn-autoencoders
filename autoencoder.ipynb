### Implementation

```python
from keras.layers import Input, Dense
from keras.models import Model
from keras.datasets import mnist
import numpy as np
```


## **Understanding the data set**
##### 1. We split the data into XTrain and YTrain and similarly Xtest and Ytest. This is given by the mnist dataset itself from Keras datasets
##### 2. We can see that the train and test data set are split as 60000 and 10000, meaning there are total of 70000 data.
##### 3. We can also see unqiue values in YTrain and YTest which are labelled from 0 to 9.


```python
#Load our MNIST dataset.
(XTrain, YTrain), (XTest, YTest) = mnist.load_data()

# shape of our dataset.
print('XTrain shape = ',XTrain.shape)
print('XTest shape = ',XTest.shape)
print('YTrain shape = ',YTrain.shape)
print('YTest shape = ',YTest.shape)
# Number of distinct values of our MNIST target
print('YTrain values = ',np.unique(YTrain))
print('YTest values = ',np.unique(YTest))
```

## **Visualization of dataset**
##### We can use matplotlib to visualize the training dataset distribution. It seems like both YTrain and YTest datasets are distributed properly so that there is very less anomaly during training.

```python
import matplotlib.pyplot as plt

# Distribution of Y dataset according to occurance vs classes in our dataset.
fig, axs = plt.subplots(1,2,figsize=(20,5)) 

axs[0].set_title('YTrain data')
axs[0].set_xlabel('Classes') 
axs[0].set_ylabel('No of data')
n, bins, patches = axs[0].hist(YTrain, ec='black')
axs[0].axhline(y=np.max(n), xmin=0, xmax=1, color='red', linestyle='--', label='MAX')

axs[1].set_title('YTest data')
axs[1].set_xlabel('Classes') 
axs[1].set_ylabel('No of data')
n, bins, patches = axs[1].hist(YTest, ec='white', color='black')
axs[1].axhline(y=np.max(n), xmin=0, xmax=1, color='red', linestyle='--', label='MAX')

plt.show()
```

## **Data Normalization as part of preprocessing**

#### Normalization
We make sure that the data are between value 0 and 1 so we divide the channel value of each by 255 which is the max color range.

#### Reshape 2D image to 1D
We perform matrix multiplication of 28 x 28 image so that resulting array is a linear 784 long.

```python
# Data normalization.
XTrain = XTrain / 255
XTest = XTest / 255

# data reshapping.
XTrain = XTrain.reshape((len(XTrain), np.prod(XTrain.shape[1:])))
XTest = XTest.reshape((len(XTest), np.prod(XTest.shape[1:])))

print (XTrain.shape)
print (XTest.shape)
```
